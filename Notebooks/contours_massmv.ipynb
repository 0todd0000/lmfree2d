## Load landmark data

Let's load the contours CSV file for the Bell dataset, which is located in the `Data` directory of this repository.

```python
dirREPO     = lm.get_repository_path()
name        = 'Bell'
fnameCSV    = os.path.join(dirREPO, 'Data', name, 'contours.csv')
r           = lm.read_csv(fnameCSV)
```

Notes:

* `dirREPO` is the repository's directory (i.e., where the repository is saved on your computer); since `dirREPO` is generally different on different computers, it is referred to generically in this notebook as `./lmfree2d/`
* `name` is the dataset name
* `fnameCSV` is the CSV file name, in this case: `./lmfree2d/Data/Bell/contours.csv`
* `r` is an array containing the contents of the contours CSV file

Let's check the contents of the `r` array.

```python
print( r.shape )
print()
print( r[0].shape )
print( r[5].shape )
print( r[7].shape )
```

Notes:

* `r` contains 10 shapes
* each element of `r` is a 2D array, representing the contour points for the given shape
* the different shapes generally have a different number of contour points

Let's plot the contours:

```python
fig,AX = plt.subplots(2, 5, figsize=(10,4))
for ax,rr in zip( AX.flatten(), r):
    ax.plot( rr[:,0], rr[:,1], '.' )
    ax.axis('equal')
plt.show()
```

The shapes can be seen clearly, but the `lmfree2d` module contains various plot functions to more conveniently visualizing the contour shapes. Here is one way:

```python
plt.figure(figsize=(10,4))
ax = plt.axes()
lm.plot_point_order(ax, r)
plt.colorbar( ax.collections[0] )
plt.show()
```

The figure above plots the same shapes, but also colors the points in the order in which they appear in the array, from the first point (blue) to the last point (red).

Note that the first and last point occur in approximately the same (homologous) location in all 10 shapes. This is generally not the case.
